## Intento integral (trapecio)

```python
import math as mt
import numpy as np

def f(x):
    return x**4 - 2*x + 1
    # return np.exp(x*x) + (1/(x*x)) + np.log(x)

N = 1000000
b = 5
a = 1
h = (b-a)/N

def A(a,b):
    return h*( (1/2)*(f(a)+f(b)) )

s = 0
for k in range(1,N):
    s = s+f(a+(k*h))

I = A(a,b)+(s*h)
I
```
