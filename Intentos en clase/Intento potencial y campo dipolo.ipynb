```python
xs = np.linspace(0.1,10,20)
ys = np.linspace(0.1,10,20)
X,Y = np.meshgrid(xs,ys)

z = np.log(Y) + np.log(X) -X -Y

pl.figure(figsize=(8,5))
pl.subplot(1,2,1)
pl.contour(xs,ys,z,levels=20)
```

```python
xs = np.linspace(0.1,10,20)
ys = np.linspace(0.1,10,20)
X,Y = np.meshgrid(xs,ys)

u = X-X*Y # f(x,y)
v = X+X*Y # g(x,y)

pl.figure(figsize=(8,8))
# pl.subplot(1,2,1)
pl.streamplot(xs,ys,u,v)
```

#### "Solución"

```python
q = 1
d = 0.5
k = 1 # Constante de Coulomb

xs = np.linspace(-1,1,100)
ys = np.linspace(-1,1,100)
X,Y = np.meshgrid(xs,ys)

# Posiciones de las cargas
r1 = np.sqrt( (X - d/2)**2 + Y**2 ) + 1
r2 = np.sqrt( (X + d/2)**2 + Y**2 ) + 1

u = q*(X - 0.25)/(r1**2) # f(x,y)
v = q*(Y - 0.25)/(r2**2) # g(x,y)

pl.figure(figsize=(8,8))
# pl.subplot(1,2,1)
pl.streamplot(xs,ys,u,v)
```
